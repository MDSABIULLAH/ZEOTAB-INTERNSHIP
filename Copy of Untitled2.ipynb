```python
# BUSINESS INSIGHT 5 :

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as pl

# Region, TransactionID, Category

# Creating the purchase matrix using crosstab
purchase_matrix = pd.crosstab(
    [complete_data['Region'], complete_data['TransactionID']],
    complete_data['Category']
)

# Get unique regions
regions = complete_data['Region'].unique()

# Initialize dictionary to store results
co_occurrence = {}

# Calculating co-occurrence for each region
for region in regions:
    # Get purchases for specific region
    region_purchases = purchase_matrix.loc[region]

    # Calculating co-occurrence matrix
    co_occurrence[region] = (
        (region_purchases.T @ region_purchases) /
        len(region_purchases)
    ).round(3)

for region in regions:
    print(f"\nCo-occurrence Matrix for {region}:")
    print(co_occurrence[region])
```

TASK - 3 [ CLUSTERING SEGMENTATION ]

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import davies_bouldin_score
import matplotlib.pyplot as plt
import seaborn as sns

data = transactions.merge(customers, on="CustomerID", how="left")

data_agg = data.groupby('CustomerID').agg({
    'TotalValue': 'sum',
    'Quantity': 'sum',
    'ProductID': 'nunique',
    'Region': 'first'
}).reset_index()

data_agg = pd.get_dummies(data_agg, columns=['Region'], drop_first=True)

scaler = StandardScaler()
data_scaled = scaler.fit_transform(data_agg.drop(columns=['CustomerID']))

num_clusters = 4
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
data_agg['Cluster'] = kmeans.fit_predict(data_scaled)

db_index = davies_bouldin_score(data_scaled, data_agg['Cluster'])

sns.pairplot(data_agg, hue='Cluster', diag_kind='kde', palette='tab10')
plt.suptitle("Customer Clusters", y=1.02)
plt.show()

data_agg.to_csv('Customer_Clusters.csv', index=False)

print(f"Number of Clusters: {num_clusters}")
print(f"Davies-Bouldin Index: {db_index:.2f}")
```
